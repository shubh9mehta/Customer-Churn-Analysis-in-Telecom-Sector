Data Preprocessing 


```python
from sklearn.preprocessing import LabelEncoder

def object_to_int(dataframe_series):
    # Check if the series is of type object(categorical data)
    if dataframe_series.dtype == 'object':
        #use LabelEncoder to convert categories to integers
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)

    return dataframe_series
```

```python
# Apply the object_to_int function to each column in the DataFrame to convert object types to integers
df = df.apply(lambda x: object_to_int(x))
df.head()
```

```python
# Calculate the correlations of all features with 'Churn'
correlation_series = df.corr()['Churn'].sort_values(ascending=False)

# Initialize the plot with specified figure size
plt.figure(figsize=(14, 7))

# Plot the correlations using a bar chart
# Since correlation_series is sorted, the bar chart will display values in sorted order
correlation_series.drop('Churn').plot(kind='bar', color='skyblue') # Drop the 'Churn' self-correlation

# Add titles and labels to make the chart informative
plt.title('Feature Correlation with Churn')
plt.xlabel('Features')
plt.ylabel('Correlation coefficient with Churn')

# Show the plot
plt.show()
```

```python
X = df.drop(columns = ['Churn'])
y = df['Churn'].values
```

```python
# Split the dataset into training and testing sets
# - 30% of data is allocated to the test set
# - Random state set to 40 for reproducible splits
# - Stratify parameter ensures class distribution in splits reflects the original dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40, stratify=y)
```

```python
def plot_distribution(feature, frame, color='r'):
    plt.figure(figsize=(8, 3))
    plt.title(f"Distribution for {feature}")
    # Using histplot for histogram and KDE
    sns.histplot(data=frame, x=feature, color=color, kde=True)
```

```python
# List of numerical column names
num_cols = ["tenure", 'MonthlyCharges', 'TotalCharges']

# Loop through each feature name and plot its distribution
for feat in num_cols:
    # Create a distribution plot for the current feature
    sns.distplot(df[feat], kde=True, hist=True)
    
    # Add a title to the plot for clarity
    plt.title(f'Distribution of {feat}')
    
    # Show the plot
    plt.show()
```

Because the numerical features are dispersed across several value ranges, I will utilize ordinary scalar scaling to bring them all into the same range

```python
df_std = pd.DataFrame(StandardScaler().fit_transform(df[num_cols].astype('float64')),
                      columns=num_cols)

# Plotting the distributions of the standardized features
plt.figure(figsize=(10, 6)) 
for feat in num_cols:
    sns.histplot(df_std[feat], color='r', kde=True, stat="density", linewidth=0)
    plt.title(f'Distribution of {feat} (Standardized)')
    plt.xlabel(f'{feat} Value')
    plt.ylabel('Density')
    plt.show()
```

```python
# Divide the columns into 3 categories, one ofor standardisation, one for label encoding and one for one hot encoding

cat_cols_ohe =['PaymentMethod', 'Contract', 'InternetService'] # those that need one-hot encoding
cat_cols_le = list(set(X_train.columns)- set(num_cols) - set(cat_cols_ohe)) #those that need label encoding
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Standardize numerical columns in the training data
X_train.loc[:, num_cols] = scaler.fit_transform(X_train[num_cols])

# Apply the same transformation to the numerical columns in the test data
X_test.loc[:, num_cols] = scaler.transform(X_test[num_cols])
```

**KNN**

```python
# Initialize the KNN model with 11 neighbors
knn_model = KNeighborsClassifier(n_neighbors=11)

# Fit the model to the training data
knn_model.fit(X_train, y_train)

# Predict on the test set
predicted_y = knn_model.predict(X_test)

# Calculate accuracy using accuracy_score for clarity
accuracy_knn = accuracy_score(y_test, predicted_y)
print(f"KNN accuracy: {accuracy_knn:.4f}")
```

```python
print(classification_report(y_test, predicted_y))
```

**svc**

```python
svc_model = SVC(random_state=1)
svc_model.fit(X_train, y_train)

# Predict on the test set
predict_y = svc_model.predict(X_test)

# Calculate accuracy using accuracy_score
accuracy_svc = accuracy_score(y_test, predict_y)

# Print the accuracy
print("SVM accuracy is:", accuracy_svc)
```

```python
print(classification_report(y_test, predict_y))
```

**Random Forest**

```python
from sklearn import metrics 

# Initialize the RandomForestClassifier 
model_rf = RandomForestClassifier(
    n_estimators=500,  # The number of trees in the forest
    oob_score=True,    # Whether to use out-of-bag samples to estimate the generalization accuracy
    n_jobs=-1,         # The number of jobs to run in parallel. -1 means using all processors
    random_state=50,   # Controls both the randomness of the bootstrapping of the samples used when building trees 
    max_features="auto",  # The number of features to consider when looking for the best split
    max_leaf_nodes=30     # Grow trees with max_leaf_nodes in best-first fashion
)

# Fit the model on the training data
model_rf.fit(X_train, y_train)

# Make predictions on the test data
prediction_test = model_rf.predict(X_test)

print(metrics.accuracy_score(y_test, prediction_test))
```

```python
print(classification_report(y_test, prediction_test))
```

```python
cm = confusion_matrix(y_test, prediction_test)

# Create the heatmap for the confusion matrix
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt="d", linecolor="k", linewidths=3, cmap="YlOrRd")  

# Add a title and adjust the fontsize
plt.title("RANDOM FOREST CONFUSION MATRIX", fontsize=14)

# Display the plot
plt.show()
```

```python
y_rfpred_prob = model_rf.predict_proba(X_test)[:, 1]
fpr_rf, tpr_rf, thresholds = roc_curve(y_test, y_rfpred_prob)

# Setup the ROC curve plot
plt.figure(figsize=(8, 6))  
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')  # Baseline for comparison
plt.plot(fpr_rf, tpr_rf, label='Random Forest', color="blue")  
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve', fontsize=16)
plt.legend(loc="best")  # Add a legend to clarify the lines
plt.grid(True)  
plt.show()
```

**Voting best classifier**

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize the classifiers
clf1 = GradientBoostingClassifier()
clf2 = LogisticRegression()
clf3 = AdaBoostClassifier()

# Create a voting classifier with soft voting
eclf1 = VotingClassifier(estimators=[
    ('gbc', clf1), 
    ('lr', clf2), 
    ('abc', clf3)
], voting='soft')

# Fit the voting classifier to the training data
eclf1.fit(X_train, y_train)

# Make predictions on the test data
predictions = eclf1.predict(X_test)

# Calculate and print the accuracy score
print("Final Accuracy Score:", accuracy_score(y_test, predictions))
```
